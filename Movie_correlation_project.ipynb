```python
# Importing Libraries

import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt

matplotlib.rcParams['figure.figsize'] = (12,8)
```

```python
df = pd.read_csv(r'C:\Users\chandresh\Downloads\movies.csv')
```

```python
df.head()
```

```python
# Lets see if there is any missing data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
print(df.isnull().sum())
```

```python
# Keep only not null values of budget and gross

df = df[df['budget'].notna()]
df = df[df['gross'].notna()]
```

```python
df.head()
```

```python
# Change data type of column

df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
```

```python
df.head()
```

```python
df['Correct_Year'] = df['released'].str.extract(r'(\d{4})')
```

```python
df.head()
```

```python
df.sort_values(by = ['gross'], inplace = False, ascending = False).head()
```

```python
pd.set_option('display.max_rows', None)
```

```python
# Drop any duplicates

df['company'].drop_duplicates().sort_values(ascending = False).head()
```

```python
plt.scatter(x = df['budget'], y = df['gross'])
plt.title('Budget vs Gross Earnings')
plt.xlabel('Budget')
plt.ylabel('Gross Earning')

plt.show()
```

```python
# Plot Budget vs Gross using seaborn

sns.regplot(x = 'budget', y = 'gross', data = df)
```

```python
# Plot Budget vs Gross using seaborn

sns.regplot(x = 'budget', y = 'gross', data = df, scatter_kws = {"color": "red"}, line_kws = {"color": "blue"})
```

```python
# Lets start looking correlation:

df.corr(method='pearson', numeric_only = True)              # Method: pearson, kendall, spearman
```

```python
correlation_matrix = df.corr(method = 'pearson', numeric_only = True)
sns.heatmap(correlation_matrix, annot = True)
plt.title('Correlation matrix for Numeric feature')
plt.xlabel('Movie Feature')
plt.ylabel('Movie Feature')

plt.show()
```

```python
df_numerized = df.copy()

for col_name in df_numerized.columns:
    if (df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
df_numerized.head()
```

```python
df.head()
```

```python
correlation_matrix = df_numerized.corr(method = 'pearson', numeric_only = True)
sns.heatmap(correlation_matrix, annot = True)
plt.title('Correlation matrix for Numeric feature')
plt.xlabel('Movie Feature')
plt.ylabel('Movie Feature')

plt.show()
```
